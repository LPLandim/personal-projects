# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

dados = pd.read_csv(url, sep = ";")
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados["Condominio"] # dados['Valor_por_mes] cria uma nova coluna contendo os dados da coluna valor + condominio
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados["IPTU"]
dados.head()
```

## Criando colunas categóricas

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'] + ' quarto(s) ' + \
                         ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quarto(s) ' + \
                         ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.' # astype() converte um dado especifico em outro tipo
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não") # .apply aplica uma determinada função em um conjunto de dados(Suites)

# lambda x = função personalizada. "x" equivale a cada linha da coluna Suites e fazendo um if se tiver ou não suites.

dados.head()
```
