 a = coeficiente angular

 n -> numero de elementos

 y = moscow
 
 x = datas 

 O coneficiente angular pode ser obitido usando a equação:

 

```python
Y = Moscow 
X = datas
n = np.size(Moscow)
```

```python
(X ** 2).shape
```

```python
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X ** 2) - np.sum(X) ** 2)
```

Coeficient Linear:



```python
b = np.mean(Y) - a * np.mean(X)

y = a * X + b
```

```python
np.linalg.norm(Moscow - y)
```
