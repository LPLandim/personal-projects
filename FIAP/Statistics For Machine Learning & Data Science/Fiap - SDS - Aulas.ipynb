```python
import math
import matplotlib.pyplot as plt

# m0 = média amostral
# s0 = sigma0 = desvio padrão amostral
# x = são os dados

def distribuicao_normal(x, m0=0, s0=1):
    raiz_dois_pi = math.sqrt(2*math.pi)
    return math.exp(-(x-m0) ** 2 / 2 / s0 ** 2) / (s0 * raiz_dois_pi)

xs = [x/10 for x in range(-50,50)]

plt.plot(xs, [distribuicao_normal(x, s0=1) for x in xs], '-', label='media=0,sigma=1')
plt.plot(xs, [distribuicao_normal(x, s0=2) for x in xs], '--', label='media=0,sigma=2')
plt.plot(xs, [distribuicao_normal(x, s0=0.5) for x in xs], '--', label='media=0,sigma=0.5')
plt.plot(xs, [distribuicao_normal(x, m0=-1, s0=1) for x in xs], '--', label='media=0,sigma=1')
plt.legend()
plt.show()
```

```python
# teste-t com uma amostra
import numpy as np
from scipy import stats

media_populacao = 10
amostra = [ 9.0345943, 11.62435047, 10.94535429, 11.53024109, 8.89014395, 9.97907435, 12.47288265,
           9.48617347, 10.88411341, 10.87918694, 10.06065414, 12.20994573, 9.55224436, 8.79477264,
           10.94354268, 9.73969884, 9.62296781, 11.98306593, 11.20673583, 12.53495058, 11.92485353]

alpha = 0.05 # representa 5% significância

estatistica_t, valor_p = stats.ttest_1samp(a=amostra, popmean=media_populacao)
print("Estatistica-t: ", estatistica_t)
print("Valor-p: ", valor_p)


# modelagem de teste de hipóteses
# H0: média == 10cm (REJEITO!)
# H1: média <> 10cm (ACEITO!)
if (valor_p < alpha):
    print("Rejeito a hipótese nula e aceito a hipótese alternativa")
    print("Logo, o tamanho médio das bolas NÃO é 10cm com significância de 5%")
else:
    print("Aceito a hipótese nula e rejeito a hipótese alternativa")
    print("Logo, o tamanho médio das bolas É 10cm com significância de 5%")
```

```python
import numpy as np
from scipy import stats

amostra_fabrica_1 = [ 9.51182903, 7.19439452, 10.42818857, 8.48623977, 9.95392617, 9.08584036, 7.88828057, 9.13283083, 9.18172261, 7.04365586, 8.57382351, 8.81922503, 6.65734872, 9.14940375, 8.73708844, 9.24066128, 9.99249732, 8.03161223, 10.90146987, 9.35144555, 8.96127198]
amostra_fabrica_2 = [11.10428279, 8.64756835, 10.3397984 , 8.89038564, 9.48686728, 10.71393526, 8.54601478, 10.58205223, 8.90051481, 9.54782322, 9.48101232, 9.22949207, 10.29398161, 11.69433187, 10.25764347, 9.06374192, 10.21169767, 9.51190852, 8.83224818, 9.57369906, 10.61416977, 12.06320768, 9.50302557, 8.82285133, 9.15629123]

estatistica_t, p_valor = stats.ttest_ind(amostra_fabrica_1, amostra_fabrica_2)
alpha = 0.05

# Testar se a media da fábrica 1 é maior do que a média da fábrica 2

# H0: média da fábrica 1 é igual ou menor a média da fábrica 2
# H1: média da fábrica 1 é maior do que a média da fábrica 2

# Erro tipo 1 é quando eu ACEITO H1, mas H1 está errado
# Erro tipo 2 é quando eu ACEITO H0, mas H0 está errado

# alpha ou nível de significância é a chance da minha modelagem
# estar incorreta, ou seja, o grau de confiança que eu tenho
# no meu teste de hipótese

# determinar qual é a minha hipótese nula
# H0: média da fabrica 1 == média da fábrica 2

# determinar qual é a hipótese alternativa
# H1: média da fabrica 1 != média da fábrica 2

# modelar o teste de hipóteses

# H0: média1 == média2
# H1: média1 != média2

print("P-valor: ", p_valor)
print("Estatística-t: ", estatistica_t)

if (p_valor < alpha):
    print("Rejeito H0, as médias são diferentes com significância de 5%")
else:
    print("Aceito H0, as médias são iguais com significância de 5%")
```

```python
import numpy as np
from scipy import stats

# Em uma fábrica, quer-se identificar se a alteração do operador da máquina,
# influencia no diâmetro médio da bola produzida

amostra_com_operador_1 = [ 9.51182903, 7.19439452, 10.42818857, 8.48623977, 9.95392617, 9.08584036, 7.88828057, 9.13283083, 9.18172261, 7.04365586, 8.57382351, 8.81922503, 6.65734872, 9.14940375, 8.73708844, 9.24066128, 9.99249732, 8.03161223, 10.90146987, 9.35144555, 8.96127198]
amostra_com_operador_2 = [11.10428279, 8.64756835, 10.3397984 , 8.89038564, 9.48686728, 10.71393526, 8.54601478, 10.58205223, 8.90051481, 9.54782322, 9.48101232, 9.22949207, 10.29398161, 11.69433187, 10.25764347, 9.06374192, 10.21169767, 9.51190852, 8.83224818, 9.57369906, 10.61416977]
alpha = 0.05

# TESTE-T DEPENDENTE

estatistica_t, p_valor = stats.ttest_rel(amostra_com_operador_1, amostra_com_operador_2)
print("P-valor: ", p_valor)

if p_valor < alpha:
    print("Rejeito H0, a troca de operador afetou o diâmetro médio com significância de 5%")
    
else:
    print("Aceito H0, a troca de operador não afetou o diâmetro médio com significância de 5%")
```

```python
# TESTE-T: HOMOCEDASTICIDADE E NORMALIDADE
# HOMOCEDASTICIDADE: A VARIÂNCIA DAS AMOSTRAS É IGUAL
# NORMALIDADE: AS OBSERVAÇÕES DA AMOSTRA SEGUEM UMA DISTRIBUIÇÃO NORMAL

# SHAPIRO TESTA A NORMALIDADE
# LEVENE TESTE A HOMOCEDASTICIDADE

ALPHA = 0.05
ESTATISTICA_T, P_VALOR = stats.shapiro(amostra_com_operador_1)

# hipótese SHAPIRO

# H0: distribuição é normal
# H1: distribuição é diferente de normal

print("p-valor: ", P_VALOR)

if P_VALOR < ALPHA:
# NÃO POSSO SEGUIR COM O TESTE-T
    print("Rejeito H0, dados não seguem a distribuição normal")
    
else:
# SOMENTE POSSO SEGUIR COM O TESTE-T, SE EU ACEITO H0 DE SHAPIRO
    print("Aceito H0, dados seguem distribuição normal")
```

```python
# LEVENE TESTE A HOMOCEDASTICIDADE
# NORMALIDADE e HOMOCEDASTICIDADE SÃO CONDIÇÕES PARA REALIZAR O TESTE-T

ALPHA = 0.05
estatistica_t, p_valor = stats.levene(amostra_com_operador_1, amostra_com_operador_2)

# hipóteses LEVENE
# H0: variância das amostras é igual
# H1: a variância das amostras é diferente

if p_valor < ALPHA:
# NÃO CONTINUO COM O TESTE-T, PARTIR PARA O TESTE DE WALCH
    print("Rejeito H0, a variância das amostras é diferente")
    
else:
# CONTINUO COM O TESTE-T
    print("Aceitar o H0, a variância das amostras é igual")
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

var1 = ['Masculino','Masculino','Masculino','Masculino','Masculino','Masculino','Feminino','Feminino','Feminino','Feminino','Feminino','Feminino']
var2 = ['Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Cachorro','Gato','Gato']

chi = pd.crosstab(var1, var2)
chi_score, p_valor, dof, expected = chi2_contingency(chi)

# chi_score: estatistica-t
# p_valor: porcentagem
# dof: graus de liberdade
# saber se as variáveis são dependentes ou não
# modelo do teste de hipóteses:
# H0 diz que não existe associação entre as variáveis (É INDEPENDENTE)
# H1 diz que existe associação entre as variáveis (É DEPENDENTE)
print("P-valor: ", p_valor)

alpha = 0.05

if p_valor < alpha:
    print("As variáveis são dependentes")
    
else:
    print("As variáveis são independentes")
```
