```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("datasets/games.csv")

df.shape
```

```python
df1 = df

df1.drop(columns=['AppID', 'Name', 'About the game', 'Header image', 'Website', 'Support email', 'Support url',
                  'Developers', 'Publishers', 'Categories', 'Genres', 'Tags', 'Screenshots', 'Movies'])
```

```python
df1 = df[(df['Price'] > 0) & (df['Median playtime forever'] > 60)]
```

```python
X = df1[['Metacritic score', 'User score', 'DLC count', 'Achievements', 'Windows', 'Mac', 'Linux', 'Average playtime forever', 'Median playtime forever']]
Y = df1['Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5, random_state=42)


model_linear = LinearRegression()
model_linear.fit(X_train, Y_train)
score_linear = model_linear.score(X_train, Y_train)


poly_features = PolynomialFeatures(degree=3)
X_poly_train = poly_features.fit_transform(X_train)
model_poly = LinearRegression()
model_poly.fit(X_poly_train, Y_train)
score_polynomial = model_poly.score(X_poly_train, Y_train)


test_score_linear = model_linear.score(X_test, Y_test)
test_score_polynomial = model_poly.score(poly_features.transform(X_test), Y_test)


print(f"R-quadrado do Modelo Linear: {score_linear}")
print(f"R-quadrado do Modelo Polinonimal: {score_polynomial}")

# Checando se tem overfitting
alpha = 0.1
t_stat, p_valor = stats.ttest_1samp([test_score_linear], popmean=score_linear)
if p_valor < alpha:
    print("Regressão Linear: O Modelo tem overfitting.")
else:
    print("Regressão Linear: O Modelo NÃO tem overfitting.")

t_stat, p_valor = stats.ttest_1samp([test_score_polynomial], popmean=score_polynomial)
if p_valor < alpha:
    print("Regressão Polinominal: O Modelo tem overfitting.")
else:
    print("Regressão Polinomial: O Modelo NÃO tem overfitting.")


```
