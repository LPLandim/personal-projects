```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm

df = pd.read_csv("imc.csv", sep = '\t')
df2 = pd.read_csv("runner.csv", sep = '\t')
df3 = pd.read_csv("visits.csv", sep = '\t')
```

```python
df2.head(10)
```

```python
# Running 
# o meu X vai ser running time e o meu Y serÃ¡ o heartbeat

# Y = B0 + B1 * X1 + E

model = LinearRegression(n_jobs = None, fit_intercept = True, copy_X = True)

x = df2["running_time"].to_numpy().reshape(-1,  1)
y = df2['heartbeat'].to_numpy()
model.fit(x, y)

# intercept= b0 = alfa =  tetha0
# coef = lista com todos os betas a partir do 1;

print(model.score(x, y))
print(model.intercept_)
print(model.coef_)
```

```python
# r_quadrado = 0.4

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

model.summary()
# Y = 285 - 3.9 * X
# r_quadrado 40%
# tenho significÃ¢ncia estatÃ­stica
```

```python
df3['date'] = pd.to_datetime(df3["date"])
df3['dow'] = df3['date'].dt.dayofweek

df3['mon'] = df3.apply(lambda row: 1 if row['dow'] == 0 else 0, axis = 1) # monday = segunda-feira
df3['tue'] = df3.apply(lambda row: 1 if row['dow'] == 1 else 0, axis = 1) # tuesday = terÃ§a-feira
df3['wed'] = df3.apply(lambda row: 1 if row['dow'] == 2 else 0, axis = 1) # wednesday = quarta-feira
df3['thu'] = df3.apply(lambda row: 1 if row['dow'] == 3 else 0, axis = 1) # thursday = quinta-feira
df3['fri'] = df3.apply(lambda row: 1 if row['dow'] == 4 else 0, axis = 1) # friday = sextafeira
df3['sat'] = df3.apply(lambda row: 1 if row['dow'] == 5 else 0, axis = 1) # saturday = sÃ¡bado
df3['sun'] = df3.apply(lambda row: 1 if row['dow'] == 6 else 0, axis = 1) # sunday = domingo

df3['dow']
```

```python
df3
```

```python
model = LinearRegression(n_jobs = None, fit_intercept = True, copy_X = True)

x = df3[['mon', 'tue','wed', 'thu', 'fri', 'sat', 'sun']].to_numpy()
y = df3['visits'].to_numpy()

model.fit(x, y)
print(model.score(x, y))

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()
```

```python
df = df.loc[df['imc'] != 0]
df = df.dropna(how= "any", axis = 0)
df
```

```python
df['drugs'] = df.apply(lambda row: 1 if row['drugs'] == 'yes' else 0, axis= 1)
df['alcohol'] = df.apply(lambda row: 1 if row['alcohol'] == 'yes' else 0, axis= 1)
df['tobacco'] = df.apply(lambda row: 1 if row['tobacco'] == 'yes' else 0, axis= 1)
df['drogas_geral'] = df.apply(lambda row: 1 if row['tobacco'] == '1' or row == 1 else 0, axis= 1)

df['sex'] = df.apply(lambda row: 1 if row['sex'] == 'Masculino' else 0, axis= 1)
df['parda'] = df.apply(lambda row: 1 if row['ethnicity'] == 'Parda' else 0, axis= 1)
df['branca'] = df.apply(lambda row: 1 if row['ethnicity'] == 'Branca' else 0, axis= 1)
df['negra'] = df.apply(lambda row: 1 if row['ethnicity'] == 'Negra' else 0, axis= 1)
df['amarela'] = df.apply(lambda row: 1 if row['ethnicity'] == 'Amarela' else 0, axis= 1)
```

```python
model = LinearRegression(n_jobs= None, fit_intercept= True, copy_X= True)

x = df[['drugs', 'alcohol', 'tobacco', 'sex', 'age', 'parda', 'branca', 'negra', 'amarela']].to_numpy()
y = df['imc'].to_numpy()
model.fit(x, y)

print(model.score(x, y))

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

print(model.summary())
```
