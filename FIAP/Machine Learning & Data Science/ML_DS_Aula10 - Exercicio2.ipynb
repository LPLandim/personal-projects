```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats

file = 'datasets/Salarios.csv'
df = pd.read_csv(file)
df
```

```python
df1 = df
df_filtered = df1.dropna()
df_filtered
```

```python
# 1- Calcular regressão para determinar a idade, filtrando nulos.
x = df_filtered[['anos de estudo', 'anos de experiencia']].to_numpy()
y = df_filtered['idade'].to_numpy()

model_idade = LinearRegression()
model_idade.fit(x, y)
score = model_idade.score(x, y)

print(f"R² da regressão com dados filtrados: {score}")
```

```python
def completar_idade(linha):
  
  if pd.isna(linha['idade']) and not pd.isna(linha['anos de estudo']) and not pd.isna(linha['anos de experiencia']):
    # Separa as variáveis regressoras utilizadas no modelo
    campos = linha[['anos de estudo', 'anos de experiencia']]
    # Retorna para o "apply" o valor previsto
    return round(model_idade.predict([campos])[0])
  # Se já tem o ano de nascimento, devolve o valor que tem
  return linha['idade']

# apply: executar a função chamada 1x para cada linha
df1['idade_calculada'] = df1.apply(completar_idade, axis=1)
df1
```

```python
# 2- Calcular regressão para determinar anos de estudo, filtrando nulos.
df_filtered = df1.dropna()

x = df_filtered[['anos de experiencia', 'idade_calculada']]
y = df_filtered['anos de estudo'].to_numpy()

model_estudo = LinearRegression()
model_estudo.fit(x, y)
score = model_estudo.score(x, y)

print(f"R² da regressão com dados filtrados: {score}")
```

```python
def completar_estudo(linha):
  
  if pd.isna(linha['anos de estudo']) and not pd.isna(linha['anos de experiencia']) and not pd.isna(linha['idade_calculada']):
    # Separa as variáveis regressoras utilizadas no modelo
    campos = linha[['anos de experiencia', 'idade_calculada']]
    # Retorna para o "apply" o valor previsto
    return round(model_estudo.predict([campos])[0])
  # Se já tem o ano de nascimento, devolve o valor que tem
  return linha['anos de estudo']

# apply: executar a função chamada 1x para cada linha
df1['estudo_calculado'] = df1.apply(completar_estudo, axis=1)
df1
```

```python
# Ajustando um modelo de regressão linear para prever 'anos de experiencia' com base em 'anos de estudo' e 'idade'
df_filtered = df1.dropna()

x = df_filtered[['estudo_calculado', 'idade_calculada']]
y = df_filtered['anos de experiencia'].to_numpy()

model_experiencia = LinearRegression()
model_experiencia.fit(x, y)
score = model_experiencia.score(x, y)

print(f"R² da regressão com dados filtrados: {score}")
```

```python
def completar_experiencia(linha):
  
  if pd.isna(linha['anos de experiencia']) and not pd.isna(linha['estudo_calculado']) and not pd.isna(linha['idade_calculada']):
    # Separa as variáveis regressoras utilizadas no modelo
    campos = linha[['estudo_calculado', 'idade_calculada']]
    # Retorna para o "apply" o valor previsto
    return round(model_estudo.predict([campos])[0])
  # Se já tem o ano de nascimento, devolve o valor que tem
  return linha['anos de experiencia']

# apply: executar a função chamada 1x para cada linha
df1['experiencia_calculado'] = df1.apply(completar_estudo, axis=1)
df1
```

```python
# 3- Recalcular a regressão original com os missings reconstruídos
x = df[['estudo_calculado', 'experiencia_calculado', 'salario']].values
y = df['idade_calculada'].values

model_idade = LinearRegression()
model_idade.fit(x, y)

score = model_idade.score(x, y)
print(f"R² da regressão com dados filtrados: {score}")
```
